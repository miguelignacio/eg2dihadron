# Q2, nu, z, central curve C, upper curve C, lower curve C, central Fe, upper Fe, lower Fe, central Pb, upper Pb and lower Pb.

```python
import csv

FF = {}

errFF_plus = {}
errFF_minus = {}


FF['C_pim'] = []
FF['Fe_pim'] = []
FF['Pb_pim'] = []


errFF_plus['C_pim'] = []
errFF_plus['Fe_pim'] = []
errFF_plus['Pb_pim'] = []

errFF_minus['C_pim'] = []
errFF_minus['Fe_pim'] = []
errFF_minus['Pb_pim'] = []

x = []

with open('clas_for_miguel_68CL_minus.txt') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        row = row[0].split()
        #print(row)
        x.append(float(row[2]))
        FF['C_pim'].append(float(row[3]))
        errFF_minus['C_pim'].append(float(row[4]))
        errFF_plus['C_pim'].append(float(row[5]))

        FF['Fe_pim'].append(float(row[6]))
        errFF_minus['Fe_pim'].append(float(row[7]))
        errFF_plus['Fe_pim'].append(float(row[8]))

        FF['Pb_pim'].append(float(row[9]))
        errFF_minus['Pb_pim'].append(float(row[10]))
        errFF_plus['Pb_pim'].append(float(row[11]))
```

```python

FF['C_pip'] = []
FF['Fe_pip'] = []
FF['Pb_pip'] = []


errFF_plus['C_pip'] = []
errFF_plus['Fe_pip'] = []
errFF_plus['Pb_pip'] = []

errFF_minus['C_pip'] = []
errFF_minus['Fe_pip'] = []
errFF_minus['Pb_pip'] = []

x = []

with open('clas_for_miguel_68CL_plus.txt') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        row = row[0].split()
        #print(row)
        x.append(float(row[2]))
        FF['C_pip'].append(float(row[3]))
        errFF_minus['C_pip'].append(float(row[4]))
        errFF_plus['C_pip'].append(float(row[5]))

        FF['Fe_pip'].append(float(row[6]))
        errFF_minus['Fe_pip'].append(float(row[7]))
        errFF_plus['Fe_pip'].append(float(row[8]))

        FF['Pb_pip'].append(float(row[9]))
        errFF_minus['Pb_pip'].append(float(row[10]))
        errFF_plus['Pb_pip'].append(float(row[11]))

```

```python
plt.errorbar(x[:10],FF['C_pim'][:10])
plt.errorbar(x[10:20],FF['C_pim'][10:20])
plt.errorbar(x[20:30],FF['C_pim'][20:30])

plt.fill_between(x[:10],errFF_minus['C_pim'][:10],errFF_plus['C_pim'][:10])

plt.errorbar(x[:10],FF['Fe_pim'][:10])
plt.errorbar(x[10:20],FF['Fe_pim'][10:20])
plt.errorbar(x[20:30],FF['Fe_pim'][20:30])

plt.fill_between(x[:10],errFF_minus['Fe_pim'][:10],errFF_plus['Fe_pim'][:10])

plt.errorbar(x[:10],FF['Pb_pim'][:10])
plt.errorbar(x[10:20],FF['Pb_pim'][10:20])
plt.errorbar(x[20:30],FF['Pb_pim'][20:30])

plt.fill_between(x[:10],errFF_minus['Pb_pim'][:10],errFF_plus['Pb_pim'][:10])

plt.ylim([0.0,1.2])
print(x[:10])
plt.show()
```

```python
plt.errorbar(x[:10],FF['C_pip'][:10])
plt.errorbar(x[10:20],FF['C_pip'][10:20])
plt.errorbar(x[20:30],FF['C_pip'][20:30])

plt.errorbar(x[:10],FF['Fe_pip'][:10])
plt.errorbar(x[10:20],FF['Fe_pip'][10:20])
plt.errorbar(x[20:30],FF['Fe_pip'][20:30])

plt.errorbar(x[:10],FF['Pb_pip'][:10])
plt.errorbar(x[10:20],FF['Pb_pip'][10:20])
plt.errorbar(x[20:30],FF['Pb_pip'][20:30])


plt.ylim([0.0,1.2])
plt.show()
```

```python
plt.figure(figsize=(8,6))

for target in ['C','Fe','Pb']:
    plt.errorbar(x[:10],FF['%s_pip'%target][:10],label='pi+ %s'%target)
    plt.errorbar(x[:10],FF['%s_pim'%target][:10],label ='pi- %s'%target)
    plt.fill_between(x[:10],errFF_minus['%s_pim'%target][:10],errFF_plus['%s_pim'%target][:10],alpha=0.3)
    plt.fill_between(x[:10],errFF_minus['%s_pip'%target][:10],errFF_plus['%s_pip'%target][:10],alpha=0.3)
    print(x[:10])
    print(FF['%s_pip'%target][:10])
    print(errFF_minus['%s_pip'%target][:10])
    print(errFF_plus['%s_pip'%target][:10])
    print('')
    plt.xlim([0.3,0.8])
    plt.ylim([0.0,1.2])

    plt.xlabel('z')
    plt.ylabel(r'$R_{h}$',fontsize=20) 
    plt.title('Calculation using nFF')
    plt.legend(loc='best',frameon=False,ncol=3,fontsize=16)
```
